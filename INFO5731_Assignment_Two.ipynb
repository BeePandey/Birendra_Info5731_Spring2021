# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
#Remove noise and special characters and punctuations

Is_CovidTweets['tweets']=Is_CovidTweets['tweets'].str.replace('[!-;:,<>./?@#$%^&*_~{}]','')
print(Is_CovidTweets)
```

```python
#Remove Numbers
Is_CovidTweets['tweets']=Is_CovidTweets['tweets'].str.replace('\d','')
print(Is_CovidTweets)
```

```python
 #Remove stopwords by using the stopwords list.
 from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
Is_CovidTweets['tweets']=Is_CovidTweets['tweets'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
print(Is_CovidTweets)
```

```python
#Lowercase All Text 

Is_CovidTweets['tweets']=Is_CovidTweets['tweets'].apply(lambda x: " ".join(x.lower() for x in x.split()))
print(Is_CovidTweets)
```

```python
#Stemming 
from nltk.stem import PorterStemmer
st=PorterStemmer()
Is_CovidTweets['tweets']=Is_CovidTweets['tweets'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
print(Is_CovidTweets)
```

```python
#Lemmatization 
import nltk
from textblob import Word
nltk.download('wordnet')
Is_CovidTweets['tweets']=Is_CovidTweets['tweets'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
print(Is_CovidTweets)
```
