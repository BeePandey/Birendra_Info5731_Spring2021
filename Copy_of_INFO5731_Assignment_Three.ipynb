```
#Scapping data and converting to dataframe

import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.common.by import By


driver = webdriver.Chrome('chromedriver',options=options)
link = 'https://www.imdb.com/title/tt10579952/reviews?ref_=tt_ov_rt'
is_arraytitle = []
is_rarray = []
driver.get(link)
for item in range(4):
  driver.find_element_by_class_name("ipl-load-more__button").click()
  time.sleep(5)
  l_title = driver.find_elements(By.CLASS_NAME, "title")
  lreviews = driver.find_elements(By.CLASS_NAME, "text")
for ele, sub_ele in zip(l_title, lreviews):
      is_arraytitle.append((ele.text).replace('\n',''))
      is_rarray.append(sub_ele.text)
dataf = pd.DataFrame(list(zip(is_arraytitle, is_rarray)), columns =['Title', 'Review'])
dataf
```

```
import nltk
nltk.download()
```

Data Cleaning and Processing 

```
#Converting to Lower Case

dataf['Lower Case'] = dataf['Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
dataf
```

```
#Removal of Punctuation

dataf['Removal of Punctuation'] = dataf['Lower Case'].str.replace('[^\w\s]','')
dataf
```

```
#Removal of Special Characters

import re
dataf['Removal of Special Characters'] = dataf['Removal of Punctuation'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
dataf
```

```
#Removal of Stop words
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
dataf['Stopwords Removal'] = dataf['Removal of Punctuation'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
print(dataf)

```

```
#Stemming

from nltk.stem import PorterStemmer
st = PorterStemmer()
dataf['After Stemming'] = dataf['Stopwords Removal'].apply(lambda x: " ".join([st.stem(word) for word in x]))
dataf
```

```
#Lemmatization

from textblob import Word
import nltk
nltk.download('wordnet')

dataf['After Lemmatization'] = dataf['After Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
dataf
dataf.to_csv('/content/sample_data/reviwes.csv',index=False)
```

1.1 - Trigrams

```
import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize


sentencesword = []
for sentence in dataf['After Lemmatization']:
  sentencesword.append(word_tokenize(sentence))
is_cleaning = [x for x in sentencesword if x != []]
iterations = list(itertools.chain.from_iterable(is_cleaning))


tri_grams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(tri_grams)
frequency_dist
```

```
# Write your code here

import math
def tf_idf_value(sentence,word):
  size = len(sentence.split(" "))
  tf_value = sentence.count(word)/size
  idf_value = 0
  if(tf_value!=0):
    idf_value = math.log(size)/sentence.count(word)
  else:
    return 0;
  return tf_value*idf_value

sentences = dataf["Stopwords Removal"].values.tolist()
tokens_list = set([j for i in sentences for j in i.split(" ")])
tf_idf = pd.DataFrame(tokens_list,columns=["token"])
count = 0
for i in sentences:
  tf_idf[str(count)] = tf_idf["token"].apply(lambda x : tf_idf_value(i,x))
  count+=1

tf_idf


```

Cosine Similarity

```
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

query_sentence = 'The perfect portrait of a decline to insanity" & Joaquin Phoenix portrays this flawlessly in a performance of a lifetime. Choreographed impeccably by Todd Phillips & is nothing short of breathtaking. The cinematography & backdrop ideology of such an original film for such a complex character is only complimented by the superb soundtrack & acting. A hauntingly beautiful film & A Definate 10/10. Edit: Bring on part 2 plssssss!. We all want it!.'
X_list = word_tokenize(query_sentence)
sw = stopwords.words('english') 
X_set = {w for w in X_list if not w in sw}
```

```
def compute_vector(X_set, Y_set):
  is_vector = X_set.union(Y_set)
  l1 =[];l2 =[]
  for w in is_vector: 
      if w in X_set: l1.append(1)
      else: l1.append(0) 
      if w in Y_set: l2.append(1) 
      else: l2.append(0)
  return is_vector, l1, l2
  
  
```

```

def compute_cosine(rvector, l1, l2):
  c = 0
  for i in range(len(rvector)): 
        c+= l1[i]*l2[i] 
  cosine = c / float((sum(l1)*sum(l2))**0.5) 
  return cosine
```

```

  
similaritem = []
for sentence in dataf['Stopwords Removal']:
  Y_list = word_tokenize(sentence)
  Y_set = {w for w in Y_list if not w in sw}
  is_vector, l1, l2 = compute_vector(X_set, Y_set)
  try:
    similarity = compute_cosine(is_vector, l1, l2)
  except ZeroDivisionError:
    similarity = 'None'
  similaritem.append(similarity)
cosine_dataframe = pd.DataFrame(list(zip(dataf['Review'],similaritem)), columns=['Reviews','Cosine Similarity'])
cosine_dataframe
```
